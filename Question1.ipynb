```python
import requests
import json
import pprint
import pandas

currency="USD"

# empty currency for getting all currencies
out_curr=""

# from date
start_date="2021-01-01"

# to date
end_date="2021-01-10"

api_url = 'https://api.exchangerate.host/timeseries?base={0}&start_date={1}&end_date={2}&symbols={3}'.format(currency,start_date,end_date,out_curr)
response = requests.get(api_url)

# retrive response in json format
data = response.json()
#print(data['rates'])

exchanges_list = list()

for date, value in data["rates"].items():
  for curr, currency_rate in value.items():
    merged_data = date +  ","+ "''"+curr +"''"+","+ str(currency_rate)
    exchanges_list.append(merged_data)

print(exchanges_list)

#pprint.pprint(data["rates"])
df = pandas.DataFrame(data={"date,name,value": exchanges_list})
df.to_csv("./file.csv", sep=',',index=False)
```
